## 7. Khám phá phân phối dữ liệu (Data Distribution)
Chúng ta sẽ vẽ biểu đồ Histogram (phân phối tần suất) và Boxplot (biểu đồ hộp) cho các thuộc tính quan trọng để xem dữ liệu tập trung ở đâu và có nhiễu (outliers) hay không.

```python
def plot_distribution(data, column, title, color='skyblue'):
    plt.figure(figsize=(12, 5))
    
    # Biểu đồ 1: Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data[column], kde=True, color=color, bins=30)
    plt.title(f'Histogram of {title}')
    plt.xlabel(title)
    plt.ylabel('Frequency')
    
    # Biểu đồ 2: Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[column], color=color)
    plt.title(f'Boxplot of {title}')
    
    plt.tight_layout()
    plt.show()

# Vẽ biểu đồ cho Nhiệt độ
plot_distribution(df, 'temperature_2m', 'Temperature (°C)', color='orange')

# Vẽ biểu đồ cho Độ ẩm
plot_distribution(df, 'relative_humidity_2m', 'Humidity (%)', color='blue')

# Vẽ biểu đồ cho Tốc độ gió
plot_distribution(df, 'wind_speed_10m', 'Wind Speed (km/h)', color='green')
```

### Nhận xét phân phối (Data Insights)
Dựa trên các biểu đồ trực quan hóa ở trên, nhóm có những nhận xét chi tiết về đặc điểm khí hậu TP.HCM trong giai đoạn thu thập:

*   **1. Nhiệt độ (`temperature_2m`):**
    *   **Phân phối:** Có dạng phân phối gần chuẩn (Normal-like) nhưng hơi lệch nhẹ. Đỉnh phổ biến nhất nằm quanh mức **26°C - 27°C**.
    *   **Khoảng giá trị:** Phần lớn dữ liệu (Interquartile Range - hộp màu vàng) tập trung ổn định trong khoảng **25°C đến 30°C**.
    *   **Ngoại lai (Outliers):** Boxplot xuất hiện nhiều điểm ngoại lai ở phía giá trị cao (**trên 37°C**). Điều này chỉ ra rằng dù nền nhiệt trung bình ổn định, vẫn xuất hiện các đợt nắng nóng cực đoan (Heatwaves).

*   **2. Độ ẩm (`relative_humidity_2m`):**
    *   **Phân phối:** Bị **lệch trái (Negative skew)** rất mạnh. Đồ thị dồn trọng số về phía bên phải.
    *   **Đặc điểm:** Độ ẩm tại TP.HCM rất cao, thường xuyên duy trì ở mức **trên 70%**. Mật độ dữ liệu dày đặc nhất ở khoảng **85% - 95%**.
    *   **Ngoại lai:** Các giá trị độ ẩm thấp (dưới 35%) xuất hiện dưới dạng outliers. Điều này cho thấy không khí khô hanh là hiện tượng rất hiếm gặp, chỉ xảy ra trong những điều kiện thời tiết đặc biệt.

*   **3. Tốc độ gió (`wind_speed_10m`):**
    *   **Phân phối:** Bị **lệch phải (Positive skew)** rõ rệt.
    *   **Đặc điểm:** Tốc độ gió chủ yếu ở mức thấp, phổ biến nhất là từ **5 - 8 km/h** (gió nhẹ).
    *   **Ngoại lai:** Có một dải dài các giá trị ngoại lai nằm ở mức cao (**trên 19 km/h**). Đây là các thời điểm có gió giật mạnh, có thể liên quan đến các cơn mưa dông hoặc ảnh hưởng của bão/áp thấp nhiệt đới.
